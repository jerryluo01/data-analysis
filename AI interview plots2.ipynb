```python
import pandas as pd
import matplotlib.pyplot as plt

movies_df = pd.read_csv("movies_data.csv")
movies_df["count"] = 1

df1 = movies_df.groupby(["rating"]).count()["count"]
df1 = df1.sort_values(ascending=False)
df2 = df1.reset_index()

ratings = df2["rating"]
counts = df2["count"]

plt.xlabel("Ratings")
plt.ylabel("Count")
plt.title("Number of movies per rating")

plt.bar(ratings, counts, width=0.5)

plt.show()
print(df1)
```
