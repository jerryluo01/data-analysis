```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

genre_df = pd.read_csv("genre_data.csv")
movies_df = pd.read_csv("movies_data.csv", usecols=["imdb_rating"])

df = pd.concat([genre_df, movies_df], axis=1)
df1 = df.groupby(["Main_Genre"])["imdb_rating"].mean()
df2 = df1.reset_index()

genres = df2["Main_Genre"]
ratings = df2["imdb_rating"]

plt.figure(figsize=(14, 7))
plt.bar(genres, ratings, zorder=2)

plt.title("Average IMDB Rating")
plt.xlabel("Main_Genre")
plt.ylabel("IMDB Rating")

plt.xticks(fontsize=8)

plt.grid(axis="y", color="white")

plt.gca().set_facecolor("lavender")

print(df2)
plt.show()
```
